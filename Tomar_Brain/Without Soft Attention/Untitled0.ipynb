```python



number_of_data = 3064
path = '/content/drive/MyDrive/Project_Brain tumor/Dataset/Data'

metadata = pd.DataFrame(columns=["ID ", "Label" ])
final_data_path = '/content/drive/MyDrive/Project_Brain tumor/Dataset/final_data'
# os.mkdir(final_data_path)
for item in range(1 , number_of_data) :
    data_file = path+ '/' + str(item)+'.mat'
    f = h5py.File(data_file,'r')

    p=Patient('','','','','')
    # p.PID=np.array(f['/cjdata/PID'])
    p.label=np.array(f['/cjdata/label'])
    # p.tumorBorder=np.mat(f['/cjdata/tumorBorder'])
    p.image = np.array(f['/cjdata/image'])
    # p.tumorMask=np.mat(f['/cjdata/tumorMask'])
    image_name= str(item)+'.jpg'
    image_name = final_data_path+'/'+ image_name
    mpimg.imsave(image_name, p.image)


    new_data = {
        'ID' : item ,

        'Label' : p.label[0][0]
    }
    metadata.loc[len(metadata)] = new_data
metadata.to_csv('/content/drive/MyDrive/Project_Brain tumor/Dataset/metadata.csv' , index=False)


```
